### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type using "groupby" and "count" on "ride_id"

total_rides_by_type = pyber_data_df.groupby(["type"]).count()["ride_id"]

total_rides_by_type
```

```python
# 2. Get the total drivers for each city type by using "groupby" and "sum" on "driver_count". Use "sum" instead of "count"
# "Count" will return the number of isntances that there is a value in the "driver_count" column. Sum will add the
# numbers together to give you the correct total.

total_drivers_by_type = city_data_df.groupby(["type"]).sum()["driver_count"]

total_drivers_by_type
```

```python
#  3. Get the total amount of fares for each city type. Use "groupby" and "sum" again here becuase we are looking for the
# total.

total_fare_by_type=pyber_data_df.groupby(["type"]).sum()["fare"]

total_fare_by_type
```

```python
#  4. Get the average fare per ride for each city type. Now that we have the totals we can find averages.

avg_fare_per_ride = total_fare_by_type / total_rides_by_type

avg_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_per_driver = total_fare_by_type / total_drivers_by_type
avg_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
summary={
    "Total Rides": total_rides_by_type,
    "Total Drivers": total_drivers_by_type,
    "Total Fares": total_fare_by_type,
    "Average Fare per Ride": avg_fare_per_ride,
    "Average Fare per Driver": avg_fare_per_driver
}
pyber_summary_df = pd.DataFrame(summary)
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df['Total Rides'] = pyber_summary_df['Total Rides'].map("{:,}".format)

pyber_summary_df['Total Drivers'] = pyber_summary_df['Total Drivers'].map("{:,}".format)

pyber_summary_df['Total Fares'] = pyber_summary_df['Total Fares'].map("${:,.2f}".format)

pyber_summary_df['Average Fare per Ride'] = pyber_summary_df['Average Fare per Ride'].map("${:,.2f}".format)

pyber_summary_df['Average Fare per Driver'] = pyber_summary_df['Average Fare per Driver'].map("${:,.2f}".format)
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date. The code is basically performing two grouping first by
# type of city then by date/time.

fare_per_day = pyber_data_df.groupby(["type","date"]).sum()["fare"]

fare_per_day
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()

#     df                   df           .reset_index()
fare_per_day = fare_per_day.reset_index()

fare_per_day.head()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

fare_per_day_pivot = fare_per_day.pivot(index='date', columns='type', values='fare')

fare_per_day_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.

janapr_fare_per_day = fare_per_day_pivot.loc['2019-01-01':'2019-04-29']

janapr_fare_per_day.tail()
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)

janapr_fare_per_day.index = pd.to_datetime(janapr_fare_per_day.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
janapr_fare_per_day.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
janapr_fare_per_week_df = janapr_fare_per_day.resample("W").sum()

janapr_fare_per_week_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

ax = janapr_fare_per_week_df.plot(figsize=(20,6), color = ["skyblue", "coral", "gold"])


#Adding a title

ax.set_title("Total Fare by City Type")

#Adding x-axis and y-axis labels
ax.set_xlabel("Month")
ax.set_ylabel("Fare ($USD)")

#Save File
plt.savefig("analysis/PyBer_fare_summary.png")

plt.show()
```
